```python
# Read data of water "super critical point"

from ase.io import read, write

# read xyz
def read_xyz(filename):
    """
    Read an XYZ trajectory file and extract atomic coordinates.
    
    Parameters:
    filename (str): Path to the XYZ file.

    Returns:
    positions (list): A list of NumPy arrays, each containing the atomic coordinates for one snapshot.
    atom_types (list): List of atomic types for each atom in the first frame.
    """
    positions = []
    atom_types = []
    
    with open(filename, 'r') as f:
        while True:
            # Read the number of atoms (first line of each frame)
            try:
                num_atoms = int(f.readline())
            except ValueError:
                break  # End of file
            
            # Skip the second line (comment line)
            f.readline()
            
            # Initialize an array to store the positions for the current frame
            frame_positions = []
            
            # Read each atom's coordinates
            for _ in range(num_atoms):
                line = f.readline().split()
                atom_type = line[0]  # Atom type (e.g., O, H)
                x, y, z = map(float, line[1:4])  # Atomic coordinates
                
                if len(positions) == 0:  # For the first snapshot, collect atom types
                    atom_types.append(atom_type)
                
                frame_positions.append([x, y, z])
            
            # Convert to NumPy array and add to the list
            positions.append(np.array(frame_positions))
    
    return positions, atom_types


wcp_positions, atoms_types = read_xyz("../../data_h2o_super_critical/vmd_traj.xyz")
```

```python
# Similarity: REMatchKernel "one snapshot" 

from dscribe.kernels import REMatchKernel

from sklearn.preprocessing import normalize
from ase.io import read
import numpy as np

#features_all = np.load("coord_desc_soap.npy")

# Before passing the features we normalize them. Depending on the metric, the
# REMatch kernel can become numerically unstable if some kind of normalization
# is not done.

#print(np.linalg.norm(a_features[0, :]))
a_features = normalize(features_ox[:, :])
#print(np.linalg.norm(a_features[0, :]))

# Calculates the similarity with the REMatch kernel and a linear metric. The
# result will be a full similarity matrix.
#re = REMatchKernel(metric="linear", alpha=1, threshold=1e-6)

# Any metric supported by scikit-learn will work: e.g. a Gaussian.
re = REMatchKernel(metric="rbf", gamma=1, alpha=1, threshold=1e-6)

atom_desc_arr = []

#for i in range(384):
#    atom_desc_arr.append(np.array([a_features[i]]))
    
for i in range(125):
    atom_desc_arr.append(np.array([a_features[i]]))

simil_metrix = np.zeros((125,125))

for i in range(125):
    for j in range(125):
        #re_kernel = re.create([atom_desc_arr[i], atom_desc_arr[j]])
        re_kernel = re.create([atom_desc_arr[i], atom_desc_arr[j]])
        simil_metrix[i][j] = re_kernel[0][1]

#print(simil_metrix)
```
